```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_frame = cv2.imread("iron_man.jpg")
```

```python
plt.imshow(img_frame)
```

```python
rgb = cv2.cvtColor(img_frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
face = classifier.detectMultiScale(img_frame)
```

```python
face = face[0]
```

```python
x, y, w, h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
output = cv2.rectangle(rgb, (x, y), (x+w, y+h), (0, 0, 255), 2)
```

```python
plt.imshow(output)
```
